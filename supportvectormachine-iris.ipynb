```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
iris = load_iris()
```

```python
# Let's create a Dataframe
df = pd.DataFrame(iris.data, columns=iris.feature_names)

df
```

```python
df['target'] = iris.target

df
```

```python
df[df.target==0].info()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
df
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], marker='*', c='green');
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], marker='*', c='red');
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```

```python
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], marker='*', c='green');
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], marker='*', c='red');
plt.xlabel('petal Length')
plt.ylabel('petal Width')
```

```python
X = df.drop(['target', 'flower_name'], axis='columns') # feature matrix
X
```

```python
y = df.target # Target vector
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
