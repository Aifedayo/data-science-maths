```python
import pandas as pd
import seaborn as sns
import numpy as np

df = pd.DataFrame({
    "company": ["Tesla", "Uber", "Amazon", "Dangote", "Jindal Steel", "Axis Oasis", "Vendatta", "Aliko", "NNPC", "Samsung", "Apple"],
    "revenue": [32.0, 23.0, 34.2, 4.5, 23.3, 12.0, 3.5, 90.0, 244.9, 345.6, 23.4]
})

df.head()
```

```python
df.plot(x="company", y="revenue", kind="bar", color="green")
```

```python
# Using log to modify smaller datasets for visibility

df.plot(x="company", y="revenue", kind='bar', logy=True);
```

```python
# Using seaborn for visualization
df = pd.read_csv(
    "/home/akeemlag/Downloads/income.csv",
    index_col=None,
    names=["income", "count"],
    skiprows=1
    )

df.head()
```

```python
sns.set_theme(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x="income", y="count", data=df, palette="rocket", hue="count")
# sns.color_palette("viridis")
g.set_xticklabels(g.get_xticklabels(),
                  rotation=45, 
                  horizontalalignment='right');
```

```python
sns.set_theme(rc={"figure.figsize": (11.7,8.27)})
g = sns.barplot(x="income", y="count", data=df, palette="flare", hue="count")
g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment="right");

g.set_xscale("log");
```
