```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("/home/akeemlag/Downloads/insurance_data.csv")
df.head()
```

```python
plt.scatter(df.age, df.bought_insurance, marker='*', color='black');
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(df[["age"]], df.bought_insurance, random_state=42, test_size=0.1) 
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
# Show the accuracy of the model
model.score(X_test, y_test)
```
