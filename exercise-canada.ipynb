```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import linear_model
```

```python
# Load the data into a DataFrame
df = pd.read_csv("/home/akeemlag/Downloads/canada_per_capita_income.csv")

df.tail(10)
```

```python
# Check if it is linear before applying Linear Regression
# Use Scatter plot by matplotlib

plt.xlabel("Year")
plt.ylabel("Income(USD)")
plt.title("Price against Year")
plt.scatter(df.year, df["per capita income (US$)"], color='red', marker='*');
plt.plot()
```

```python
# Let's fit our model for training using linear regression
reg = linear_model.LinearRegression()
```

```python
reg.fit(X=df[["year"]], y=df["per capita income (US$)"])
```

```python
reg.predict([[2020]])
```
